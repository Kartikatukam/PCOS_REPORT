```python
import matplotlib.pyplot as plt
import seaborn as sns

# Convert object columns to numeric if possible, replacing errors with NaN
data['II    beta-HCG(mIU/mL)'] = pd.to_numeric(data['II    beta-HCG(mIU/mL)'], errors='coerce')
data['AMH(ng/mL)'] = pd.to_numeric(data['AMH(ng/mL)'], errors='coerce')

# Summary statistics for numeric columns
summary_stats = data.describe()

# Visualize the distribution of BMI and its relation to PCOS
plt.figure(figsize=(12, 6))

# BMI Distribution
plt.subplot(1, 2, 1)
sns.histplot(data['BMI'], kde=True, color='blue', bins=20)
plt.title('BMI Distribution')
plt.xlabel('BMI')
plt.ylabel('Frequency')

# BMI vs PCOS
plt.subplot(1, 2, 2)
sns.boxplot(x='PCOS (Y/N)', y='BMI', data=data, palette='Set2')
plt.title('BMI vs PCOS (Y/N)')
plt.xlabel('PCOS (Y/N)')
plt.ylabel('BMI')

plt.tight_layout()
plt.show()

summary_stats
```

BMI Distribution:

The BMI values are concentrated around 24, with a mean of ~24. This suggests most individuals fall into the "normal" to "overweight" range.

BMI vs PCOS:

Individuals with PCOS tend to have slightly higher BMI values, as seen in the box plot.

```python
import matplotlib.pyplot as plt
import seaborn as sns

symptom_columns = ['Pimples(Y/N)', 'Skin darkening (Y/N)', 'Hair loss(Y/N)']
symptom_data = data[symptom_columns + ['PCOS (Y/N)']]

symptom_melted = symptom_data.melt(id_vars=['PCOS (Y/N)'], var_name='Symptom', value_name='Presence')

plt.figure(figsize=(14, 8))
sns.countplot(data=symptom_melted[symptom_melted['Presence'] == 1], x='Symptom', hue='PCOS (Y/N)', palette='Set1')
plt.title('Symptom Prevalence by PCOS Status')
plt.xlabel('Symptom')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend(title='PCOS (Y/N)', loc='upper right')
plt.tight_layout()
plt.show()
```

Common Symptoms in PCOS:

Irregular Cycle: Often linked with hormonal imbalances, irregular menstrual cycles are a prevalent symptom among individuals with PCOS.

Pimples (Acne): Elevated androgen levels can cause acne, observed frequently in PCOS patients.

Hair Growth (Hirsutism): Excessive facial and body hair growth is a common symptom due to increased androgen levels.

Skin Darkening (Acanthosis Nigricans): Insulin resistance, a common issue in PCOS, may lead to darkened skin patches.

Hair Loss: Androgenic alopecia, or hair thinning, is often seen among PCOS patients.
